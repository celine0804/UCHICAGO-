```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np 
```

```python
iowa_df = pd.read_csv('State_of_Iowa_Salary_Book_20241030.csv')
```

```python
iowa_df.head(10)
```

```python
#data cleaning 1: removing the commas in Base Salary category 
iowa_df['Base Salary'] = iowa_df['Base Salary'].str.replace(',', '')
iowa_df.sample(10)
```

```python
#data cleaning2: removing dollar signs 
iowa_df['Base Salary'] = iowa_df['Base Salary'].str.replace('$', '')
iowa_df.sample(10)
```

```python
#data cleaning 3: delete rows of people who had their jobs terminated
iowa_df = iowa_df[iowa_df['Base Salary'] != 'TERMINATED']
iowa_df.sample(10)
```

```python
nan_count = iowa_df['Agency/Institution'].isna().sum()
nan_count
```

```python
#since there are so many missing entries for the Agency/Institution, I will remove the following row. 
iowa_df = iowa_df.drop(columns=['Agency/Institution'])
iowa_df.sample(10)
```

```python
iowa_df.columns
```

```python
iowa_df.describe()
```

Through the .describe() function, we can see that there is missing data in the Total Salary Paid column and Travel& Subsistence column, as their salary cannot be in a negative number. 

```python
#imputing 0 to rows where either the total salary paid or travel & subsistence is negative 
iowa_df.loc[iowa_df['Total Salary Paid'] < 0, 'Total Salary Paid'] = 0
iowa_df.loc[iowa_df['Travel & Subsistence'] < 0, 'Travel & Subsistence'] = 0
```

```python
iowa_df.describe()
```

```python
# Adding a new column called "Total Salary with Travel," as it's important to see how much they earned in total, including the travel and subsistence. 
iowa_df['Total Salary with Travel'] = iowa_df['Total Salary Paid'] + iowa_df['Travel & Subsistence']
iowa_df.describe()
```

```python
#relationship of the average total salary throughout the years 
average_salary = iowa_df.groupby('Fiscal Year')['Total Salary Paid'].mean().reset_index()

# line plot to visualize the average total salary by fiscal year
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_salary, x='Fiscal Year', y='Total Salary Paid', marker='o')
plt.title('Average Total Salary by Fiscal Year (Travel & Subsistence Excluded)')
plt.xlabel('Fiscal Year')
plt.ylabel('Average Total Salary Paid')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.grid()
plt.show()
```

Through the line chart, we can find out that the average total salary has been gradually increasing since 2008. 

```python
#relationship of the average total salary throughout the years 
average_salary = iowa_df.groupby('Fiscal Year')['Total Salary with Travel'].mean().reset_index()

# line plot to visualize the average total salary by fiscal year
plt.figure(figsize=(12, 6))
sns.lineplot(data=average_salary, x='Fiscal Year', y='Total Salary with Travel', marker='o')
plt.title('Average Total Salary by Fiscal Year (Travel & Subsistence Included)')
plt.xlabel('Fiscal Year')
plt.ylabel('Average Total Salary Paid')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.grid()
plt.show()
```

However, if 'travel and subsistence' column is included in the overall salary, we can find a huge decline in the average total salary from in 2010, which gradually increases after.

```python
#relationship between the gap of average salary between man and woman 
average_salary = iowa_df.groupby(['Fiscal Year', 'Gender'])['Total Salary Paid'].mean().reset_index()

# Pivot the data to make it easier to plot
salary_pivot = average_salary.pivot(index='Fiscal Year', columns='Gender', values='Total Salary Paid')

# Calculate the gap between men and women's average salary
salary_pivot['Salary Gap'] = salary_pivot['M'] - salary_pivot['F']

# Step 3: Create the line plot for average salaries and the gap
plt.figure(figsize=(14, 7))

# Plotting average salaries for men and women
sns.lineplot(data=salary_pivot, x=salary_pivot.index, y='M', label='Men Average Salary', marker='o')
sns.lineplot(data=salary_pivot, x=salary_pivot.index, y='F', label='Women Average Salary', marker='o')

# Plotting the salary gap
sns.lineplot(data=salary_pivot, x=salary_pivot.index, y='Salary Gap', label='Salary Gap (Men - Women)', marker='o', linestyle='--')

# Adding titles and labels
plt.title('Average Salary Gap Between Men and Women Throughout Fiscal Years')
plt.xlabel('Fiscal Year')
plt.ylabel('Average Salary')
plt.xticks(rotation=45)  # Rotate x-axis labels if necessary
plt.grid()
plt.legend()
plt.tight_layout()  # Adjust layout to avoid clipping
plt.show()
```

This line plot shows the difference in average salary between men and women in the same fiscal year, as well as the average salary gap. The data reveals that the average salary gap between men and women has remained stable and simiar for the past 15 years, which is around $10000 per year.  

```python
#figuring out how many people work on what department 
employee_count_by_department = iowa_df['Department'].value_counts().reset_index()
employee_count_by_department.columns = ['Department', 'Number of Employees']

plt.figure(figsize=(12, 6))
sns.barplot(data=employee_count_by_department, x='Department', y='Number of Employees', palette='viridis')
plt.title('Number of Employees by Department')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)  
plt.grid(axis='y') 
plt.show()
```

Through this chart, we can figure out that most of the respondents of this data work for either the University of Iowa or Iowa State University. However, to get a closer look on the others, I would like to present a new barplot that excludes people that work from the University of Iowa or Iowa State University. 

```python
filtered_df = iowa_df[~iowa_df['Department'].isin(['University of Iowa', 'Iowa State University'])]

# Step 2: Count the number of people in each remaining department
employee_count_filtered = filtered_df['Department'].value_counts().reset_index()
employee_count_filtered.columns = ['Department', 'Number of Employees']

# Step 3: Create the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=employee_count_filtered, x='Department', y='Number of Employees', palette='viridis')
plt.title('Number of Employees by Department (Excluding University of Iowa and Iowa State University)')
plt.xlabel('Department')
plt.ylabel('Number of Employees')
plt.xticks(rotation=90)  
plt.grid(axis='y')  
plt.show()
```

```python
#figuring out the average salary difference between the department 

average_salary_by_department = iowa_df.groupby('Department')['Total Salary Paid'].mean().reset_index()

# Step 2: Create the bar chart
plt.figure(figsize=(12, 6))
sns.barplot(data=average_salary_by_department, x='Department', y='Total Salary Paid', palette='viridis')
plt.title('Average Salary by Department')
plt.xlabel('Department')
plt.ylabel('Average Total Salary Paid')
plt.xticks(rotation=90)  
plt.grid(axis='y')   
plt.show()
```

Through this bar chart, we were able to find out that people who work on the Office of Chief Information and the Department of Management tend to earn the most money, while people who work on the Office of Governor-Elect tends to make the least money per year.  

```python
#Is Office of Governer-Elect an outlier? 
governor_elect_data = iowa_df[iowa_df['Department'] == 'Governor-Elect, Office of']

# Display the resulting DataFrame
print(governor_elect_data)
```

```python
plt.figure(figsize=(10, 6))
sns.scatterplot(data=iowa_df, x='Base Salary', y='Total Salary Paid', alpha=0.7)
plt.title('Scatter Plot of Base Salary vs. Total Salary Paid')
plt.xlabel('Base Salary')
plt.ylabel('Total Salary Paid')
plt.grid(True)
plt.tight_layout()
plt.show()

correlation_coefficient = iowa_df['Base Salary'].corr(iowa_df['Total Salary Paid'])
print(f'Correlation Coefficient between Base Salary and Total Salary Paid: {correlation_coefficient:.2f}')
```
